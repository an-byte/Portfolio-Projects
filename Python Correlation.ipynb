## Importing Libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

## Reading in the data

```python
df = pd.read_csv(r'C:\Users\anant\Desktop\Personal Projects\Python - Correlation Project\movies.csv')
```

```python
df.head()
```

## Data Cleaning

```python
# Checking for missing values

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Drop the rows with missing values

df = df.dropna()
```

```python
df.dtypes
```

```python
#Changing Data Types of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.dtypes
```

According to the documentation provided with the given dataset, the year given as a part of the date in the 'released' column is not the same as the year mentioned in the 'year' column for a few movies. In order to fix this, we will create a new column and fill it by extracting the year from the dates given in the 'released' column.

```python
df['year_corrected'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df
```

```python
# Order data by Gross Revenue

df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Dropping duplicates

df.drop_duplicates()
```

## Correlation of different features in the dataset

The main aim of the project is to check how different features of a movie impact its gross earnings. We will do this by plotting scatterplots, regression plots and heatmaps. 

```python
# Creating scatterplot of budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.title('Budget vs Gross')
plt.show()
```

```python
# Creating a regression plot for budget vs gross

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color":"red"}, line_kws = {"color":"blue"})
```

```python
# Generating Correlation Matrix
# There are three types of correlation. By default, we use 'Pearson' correlation. The others are 'Kendall' and 'Spearman'.
df.corr()
```

```python
# Creating a heatmap of the correlation matrix to better visualise its results

correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

To check for the correlation between the company that releases the movie and the amount of money it generates, we must first convert the 'company' column into numeric form.

```python
df_numeric = df

for col_name in df_numeric.columns:
    if(df_numeric[col_name].dtype == 'object'):
        df_numeric[col_name] = df_numeric[col_name].astype('category')
        df_numeric[col_name] = df_numeric[col_name].cat.codes
```

```python
df_numeric.head()
```

As we can see above, the names of the company have been replaced with a randomly generated code.

```python
correlation_matrix = df_numeric.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
df_numeric.corr()
```

In order to make the above correlation matrix easier to read and interpret, we start off by unstacking it and follow this by sorting the correlation pairs.

```python
corr_matrix = df_numeric.corr()
corr_pairs = corr_matrix.unstack()
corr_pairs
```

```python
#Sorting

corr_pairs_sorted = corr_pairs.sort_values()
corr_pairs_sorted
```

To check which features have high correlation, we can return pairs from 'corr_pairs_sorted' where the correlation value is  greater than 0.5.

```python
high_correlation = corr_pairs_sorted[(corr_pairs_sorted) > 0.5]
high_correlation
```
